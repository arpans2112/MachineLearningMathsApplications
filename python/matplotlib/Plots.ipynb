# Bar Chart

```python
import matplotlib.pyplot as plt

d = {'A': 10, 'B': 25, 'C': 15, 'D': 30}

plt.bar(range(len(d)), list(d.values()), align='center', color=['red', 'green', 'blue', 'orange'])
plt.xticks(range(len(d)), list(d.keys()))

plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Bar Chart Example')

plt.show()
```

```python
import matplotlib.pyplot as plt

d1 = {'A': 10, 'B': 25, 'C': 15, 'D': 30}
d2 = {'A': 15, 'B': 20, 'C': 12, 'D': 28}

# Plot the first set of bars
plt.bar(range(len(d1)), list(d1.values()), align='center', color=['red', 'green', 'blue', 'orange'], label='Set 1')

# Plot the second set of bars with different colors
plt.bar(range(len(d2)), list(d2.values()), align='edge',  color=['cyan', 'magenta', 'yellow', 'purple'], label='Set 2')

plt.xticks(range(len(d1)), list(d1.keys()))

plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Bar Chart Example')
plt.legend()

plt.show()
```

### Plot two charts side by side

```python
import numpy as np
import matplotlib.pyplot as plt

d1 = {'A': 10, 'B': 25, 'C': 15, 'D': 30}
d2 = {'A': 15, 'B': 20, 'C': 12, 'D': 28}

# Get the positions for the bars
x = np.arange(len(d1))

# Width of the bars
width = 0.35

# Plot the first set of bars
plt.bar(x - width/2, list(d1.values()), width, align='center', color='red', label='Set 1')

# Plot the second set of bars
plt.bar(x + width/2, list(d2.values()), width, align='center', color='blue', label='Set 2')

# Set the tick positions and labels
plt.xticks(x, d1.keys())

plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Bar Chart Example')
plt.legend()

plt.show()
```

### Plot n bars side by side

```python
import numpy as np
import matplotlib.pyplot as plt

# Sample data for n sets of bars
data = {
    'Set 1': {'A': 10, 'B': 25, 'C': 15, 'D': 30},
    'Set 2': {'A': 15, 'B': 20, 'C': 12, 'D': 28},
    'Set 3': {'A': 8, 'B': 18, 'C': 10, 'D': 25},
}

# Number of sets (n)
n = len(data)

# Get the positions for the bars
x = np.arange(len(data['Set 1']))

# Width of each bar
width = 0.2

# Plot the bars for each set
for i, (label, values) in enumerate(data.items()):
    plt.bar(x + (i - (n-1)/2) * width, list(values.values()), width, align='center', label=label)

# Set the tick positions and labels
plt.xticks(x, data['Set 1'].keys())

plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Bar Chart Example')
plt.legend()

plt.show()
```
