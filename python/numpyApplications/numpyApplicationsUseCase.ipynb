## Frobenius Norm.  





```python
def getFrobeniusNorm(numpyArray):
    '''
    numpyArray : is numpy array for which you want to get the FrobeniusNorm. 
    @return : a FrobeniusNorm value
    '''
    return np.sqrt(np.sum(np.square(numpyArray)))
```

```python
a = np.array([[2,2],[2,2]])
print(getFrobeniusNorm(a))
```

## Cosine Similarity.  

```python
def cosine_similarity(A, B):
    '''
    Input:
        A: a numpy array which corresponds to a word vector
        B: A numpy array which corresponds to a word vector
    Output:
        cos: numerical number representing the cosine similarity between A and B.
    '''
    # you have to set this variable to the true label.
    cos = -10
    dot = np.dot(A, B)
    norma = np.linalg.norm(A)
    normb = np.linalg.norm(B)
    cos = dot / (norma * normb)

    return cos
```

## Convert Degrees to Radian

```python
def degreeToRadian(degree):
    '''
    Input:
      degree: takes the degree angle and convert back to Radian angle, 
    
    Output:
      radian: We need that since numpy functions use the Radian angle for the calculation
    '''
    return degree * (np.pi / 180) # Convert degrees to radians
```

```python
print(degreeToRadian(100))
print(np.cos(degreeToRadian(0)))
print(np.sin(degreeToRadian(0)))
```
